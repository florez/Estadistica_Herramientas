```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
theta = np.linspace(0,3*np.pi*2,100)
```

```python
def Getr(theta,b):
    return b*theta
```

```python
x = Getr(theta,1.0)*np.cos(theta)
y = Getr(theta,1.0)*np.sin(theta)
```

```python
fig = plt.figure( figsize=(10,5) )
ax = fig.add_subplot(111)
#ax.axis('equal')
ax.set_xlabel(r'$x$',fontsize=15)
ax.set_ylabel(r'$y$',fontsize=15)
ax.plot(x,y)
plt.savefig('Espiral.pdf')
```
