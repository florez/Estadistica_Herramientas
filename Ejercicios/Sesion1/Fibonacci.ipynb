```python
import numpy             as np
import matplotlib.pyplot as plt
```

```python
aureo = (1 + np.sqrt(5))/2

def ComputeSerie(N):
    
    serie = np.array([])
    aureo_ = np.array([])
    a,b = 0,1
    
    serie = np.hstack([a,b])
    
    while b < N:    
        a,b = b, a+b
        serie = np.append(serie,b)
        aureo_ = np.append( aureo_, float(b)/float(a) )
    
    return serie,aureo_
```

```python
serie,aureo_ = ComputeSerie(5000)
len(serie)
```

```python
serie[:20]
```

```python
fig = plt.figure( figsize=(10,5) )
ax = fig.add_subplot(121)
ax1 = fig.add_subplot(122)
ax.plot(serie,color='k',label='Serie Fibonacci')
ax1.plot(aureo_,label='Estimación usando la serie')
ax1.axhline(y = aureo, ls='--', color='r',label='Numero Aureo')
ax.legend(loc=0)
ax1.legend(loc=0)
plt.savefig('Fibonacci.pdf')
```
