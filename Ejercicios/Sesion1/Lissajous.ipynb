```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(10,10))
theta = np.linspace(0,2*np.pi,100)
```

```python
k = 1
nside = 5
for i in range(nside):
    for j in range(nside):
        fig = plt.subplot(nside,nside,k)
        k+=1
        x = np.sin((i+1)*theta)
        y = np.sin((j+1)*theta + 0*np.pi/4.)
        
        if i <= j:
            plt.plot(x,y)
        
        plt.axis('off')
        
plt.savefig('Lissajous_0.pdf')
```
